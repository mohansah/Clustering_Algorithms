```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import random
```

```python
# Reading of file
def ReadFile(file):
    f=open(file,"r")
    lines=f.read().splitlines()
    f.close()

    items=[]
    for i in range(len(lines)):
        line=lines[i].split(',')
        feature=[]
        for j in line:
            feature.append(float(j))

        items.append(feature)
        
    return items
```

```python
def Figure(items):
    plt.figure(figsize=(20,10))
    for item in items:
        plt.scatter(item[0],item[1],c='b',s=70)
    plt.title('Before applying K Means')
    plt.show()
```

```python
def FinalFigure(items,clusters):
    plt.figure(figsize=(20,10))
    for i in range(len(items)):
        colors=['k','b','y','g','r']
        plt.scatter(items[i][0],items[i][1],c=colors[clusters[i]-1],s=70)
    plt.title('After applying K Means')
    plt.show()
```

```python
def PearsonCorrelation(obj1,obj2):
    obj1bar=0
    obj2bar=0
    for j in range(len(obj1)):
        obj1bar+=obj1[j]
        obj2bar+=obj2[j]
    obj1bar=obj1bar/float(len(obj1))
    obj2bar=obj2bar/float(len(obj2))
    up=0
    dn1=0
    dn2=0
    for i in range(len(obj1)):
        up+=(obj1[i]-obj1bar)*(obj2[i]-obj2bar)
        dn1+=math.pow((obj1[i]-obj1bar),2)
        dn2+=math.pow((obj2[i]-obj2bar),2)
    if(up!=0):    
        return up/(math.sqrt(dn1*dn2))
    else:
        return 1
```

```python
def Correlation(items,means):
    k=len(means)
    correlations=[]
    for i in range(k):
        correlation=[]
        for item in items:
            cor=PearsonCorrelation(item,means[i])
            correlation.append(cor)
        correlations.append(correlation)
    return correlations
```

```python
def InitializeMeans(items, k):    
    no_f = len(items[0]); 
    means =[]
    while(len(means)!=k):
        mean=list(random.choice(items))
        if mean not in means:
            means.append(mean)
    return means
```

```python
def CalculateMean(items,clusters,i,j):
    sum=0.0
    count=0
    z=len(items)
    for k in range(z):
        if(clusters[k]==j+1):
            sum+=items[k][i]
            count+=1
    if(count!=0):
        return(sum/float(count))
    else:
        return -1
```

```python
def UpdateMean(items,clusters,means):
    a=len(means)
    b=len(means[0])
    for j in range(a):
        for i in range(b):
            num=CalculateMean(items,clusters,i,j)
            if(num!=-1):
                means[j][i]=num
    return means
```

```python
def ClusterAssigning(clusters,Correlations):
    for j in range(len(clusters)):
        max=-2
        for i in range(len(Correlations)):
            if max<Correlations[i][j]:
                max=Correlations[i][j]
                clusters[j]=i+1
    return clusters
```

```python
def KMeans(file,no_clus,num_of_iterations):
    items=ReadFile(file)
    Figure(items)
    clusters=[0 for i in range(len(items))]
    means=InitializeMeans(items,no_clus)
    while(num_of_iterations):
        Correlations=Correlation(items,means)
        clusters=ClusterAssigning(clusters,Correlations)
        #print("---------------Means---------------------------")
        #print(means)
        means=UpdateMean(items,clusters,means)
        #print("---------------Clusters---------------------------")
        #print(clusters)
        #print()
        num_of_iterations-=1
    FinalFigure(items,clusters)
    print('Points\t\tCluster')
    print('----------------------')
    for i in range(len(items)):
        print(items[i],'\t',clusters[i])
```

```python
# Parameters : ( File_Name, No_of_Clusters, No_of_Iterations )
KMeans('sample.txt',4,5)
```
